```python
## Drop unnecessary columns
df=df.drop(['instant','dteday','casual', 'registered','atemp'], axis=1)
df.head()
```

## Step 2: Preparing the Data for Modelling
* Encoding
    * Converting binary vars into 0/1
    * Other categorical vars into dummy vars
* Splitting data into train and test data
* Rescaling the variables

## Dummy Variables

```python
df_new = df ## Creating a new dataframe to operate on
```

```python
months = pd.get_dummies(df_new['mnth'], drop_first=True)
weekday = pd.get_dummies(df_new['weekday'], drop_first=True)
weathersit = pd.get_dummies(df_new['weathersit'], drop_first=True)
season = pd.get_dummies(df_new['season'], drop_first=True)
```

```python
## Join the dummy data frame with original dataframe
df_new = pd.concat([df_new,months,weekday,weathersit,season],axis=1)
df_new.head()
```

```python
## Dropping season, weekday,weathersit, mnth
df_new.drop(['season', 'weekday','weathersit', 'mnth','yr'], axis=1, inplace=True)
df_new.head()
```

```python
df_new.drop(['yr'], axis=1, inplace=True)
```

```python
df_new.shape
```

```python
## Heat Map for analysing the co-relation between different variables
plt.figure(figsize=(25,20))
sns.heatmap(df_new.corr(), annot=True, cmap='Spectral')
plt.title("Correlation between Variables")
plt.show()
```

## Splitting data into train and test

```python
df_train, df_test = train_test_split(df_new, train_size=0.7, random_state=100)
print(f"Training Data 70% : {df_train.shape}")
print(f"Test Data 30% {df_test.shape}")
```

#### Bringing all the numerical variables to one scale so that model can process & interpret accordingly for better results.
* Therefore, using MinMaxScale() technique.

```python
## Intstantiate object of class MinMaxScaler()
scaler = MinMaxScaler()

## Create list of numeric values
num_vars = ['temp', 'hum', 'windspeed', 'cnt']

## Fit and transform the data
df_train[num_vars] = scaler.fit_transform(df_train[num_vars])
df_train.head()
```

## Step 3: Training the Model

#### The type of question that we will answer with these models are:
* what are the challenges we trying to address using this dataset?
* how many variable out these will support our model?
* which strategy will be help make the model more robust, rfe, adding one by one variables?

### Create co-relation using HeatMap

```python
plt.figure(figsize=(25,20))
sns.heatmap(df_train.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation between variables on the TRAINING Data set (70%)')
plt.show()
```

#### Dividing the dataset in X & Y for modelling

```python
X_train = df_train
```

```python
y_train = df_train.pop('cnt')
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
# add constant to the statsmodel as it does not come by default. And here we are adding it for the 1 variable i.e temp
X_train_sm = sm.add_constant(X_train['temp'])

# create 1st Model
lr = sm.OLS(y_train, X_train_sm)

# Try to do the best fit
lr_model = lr.fit()

# Get the parameters: So here, the constant & co-efficent of area is +ve - therefore, it is good interpretation - the area goes up the price will go up
lr_model.params

# We just do not want to rely on the co-efficent just - thereforem, we will also check the other parameter statistics like p-value, F-statistics using summary
lr_model.summary()
```

### Few Interpretation for considering variable=temp
#### Answer: 
* The p-value is 0, therefore, it is a good sign.
* R2 value is ~42% which means that when take into account temp as one of the variable - there is ~42% variace in the price of the house
* So, conclusion is: it is not bad - but we need some more variable to make decision based on it. Chances are it may not be significant as temp or vice versa.

#### Lets, add some more variable to our model (as show below). This method that we applying is forward selection technique

```python
# add another variable (choice is random you can select any variable) we will select bathrooms
X_train_sm = sm.add_constant(X_train[['temp','hum']])
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.params
lr_model.summary()
```

###  Interpreatation of above summary for new variable (temp,hum)?
* Again the p-values are 0 for all the co-efficents - that is significant that it is a good sign
* R2 values is ~44% - which means that there is ~44% variance in the count of rental bikes when we consider temp & hum as part of the prediction
* So, we have arrieved at the value which ~50%, however, there is no significant improvement in the R2 value. Its only 2% more, which does not have that impact.

#### Lets look at the behaviour of the model how it predicts if we add more variable

```python
# add another variable (choice is random you can select any variable) we will select bathrooms
X_train_sm = sm.add_constant(X_train[['temp','hum','windspeed']])
lr = sm.OLS(y_train, X_train_sm)
lr_model = lr.fit()
lr_model.params
lr_model.summary()
```

#### Again the same conclusion: as there is no significant improvement in the R2 values

### Now since adding the variables will be long haul, we should make use of another technique called RFE

```python
# Running RFE with the output number of the variable equal to 10
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

```python
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

### Building model using statsmodel, for the detailed statistics

```python
# Creating X_test dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
X_train_rfe.dtypes
```

```python
X_train_rfe.head()
```

```python
lm = sm.OLS(y_train,X_train_rfe)   # Running the linear model
```

```python
lr_model = lm.fit()
```

## Calculate VIF

```python
## Dropping the const variable
X_train_new = X_train_rfe.drop(['const'], axis=1)
```

```python
vif = pd.DataFrame()
X = X_train_new
vif['Features'] = X.columns
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

## Step 4: Residual Analysis of the train data

```python
X_train_lm5
```

```python
X_train_lm5=sm.add_constant(X_train_lm5)
```

```python
y_train_count = lm5.predict(X_train_lm5)
```
